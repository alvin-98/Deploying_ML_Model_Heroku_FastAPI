```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "./data/census.csv"
df = pd.read_csv(filename)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.columns = [col.strip() for col in df.columns]
```

```python

continous_features = set(df.describe().columns)
categorical_features  = list(set(df.columns) - continous_features)

df[categorical_features] = df[categorical_features].applymap(str.strip)
```

```python
[df[cat].unique() for cat in categorical_features]
```

```python
df.replace('?', np.NaN, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.to_csv('data/clean_data.csv', index=False)
```
